## Introduction
In this task, we will explore and visualize the **Iris dataset** using Python. The Iris dataset is a classic dataset in machine learning and statistics, often used for demonstrating data visualization techniques and classification algorithms.

The **Iris dataset** is one of the most famous and widely used datasets in data science and machine learning. It contains information about **150 iris flowers**, collected from three different species:

- *Iris setosa*  
- *Iris versicolor*  
- *Iris virginica*

Each flower is described using **four features** (measured in centimeters):

1. **Sepal Length**  
2. **Sepal Width**  
3. **Petal Length**  
4. **Petal Width**

The dataset is often used for **classification tasks**, **visualization**, and **exploratory data analysis** (EDA) because it is clean, small, and easy to understand.

## Importing Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

## Load the Dataset

```python
df=sns.load_dataset('iris')
```

## Create Basic Visualizations

### Scatter Plots:

#### Petal Length vs Petal Width


```python
sns.scatterplot(data=df, x="petal_length", y="petal_width", hue="species")
plt.title("Scatter Plot of Petal Length vs Petal Width")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.grid(True)
plt.show()
```

#### Petal Length vs Sepal Width

```python
sns.scatterplot(data=df, x="petal_length", y="sepal_width", hue="species")
plt.title("Scatter Plot of Petal Length vs Sepal Width")
plt.xlabel("Petal Length (cm)")
plt.ylabel("sepal Width (cm)")
plt.grid(True)
plt.show()
```

### Histograms of Iris Dataset Features

The plots below display the distribution of the four main features of the Iris dataset:

- **Sepal Length**
- **Sepal Width**
- **Petal Length**
- **Petal Width**

Each histogram shows how the values are distributed for that feature. The **KDE curve** (blue line) gives an estimate of the probability distribution. These plots help in understanding data spread, skewness, and patterns.


```python
sns.set(style="whitegrid")
# List of feature names
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
plt.figure(figsize=(12, 8))
# Loop to create histograms
for i, feature in enumerate(features):
    plt.subplot(2, 2, i+1)
    sns.histplot(data=df, x=feature, kde=True, color='skyblue')
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature.capitalize())
    plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

### Box Plots of Iris Dataset Features

The box plots below visualize the **spread**, **central tendency**, and **potential outliers** for each numerical feature in the Iris dataset:

- **Median** is shown by the line inside the box.
- **Interquartile Range (IQR)** is represented by the box.
- **Whiskers** extend to show the range, excluding outliers.
- **Outliers** are shown as individual dots beyond the whiskers.

These plots help in understanding the **distribution symmetry**, **spread**, and identifying any **unusual data points**.


```python
plt.figure(figsize=(12, 8))
# List of features to plot
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
# Create box plots for each feature
for i, feature in enumerate(features):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(data=df, y=feature, color='lightgreen')
    plt.title(f'Box Plot of {feature}')
    plt.ylabel(feature.capitalize())
    
plt.tight_layout()
plt.show()
```
